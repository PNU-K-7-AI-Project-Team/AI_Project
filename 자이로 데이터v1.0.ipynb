### 데이터 시각화
> 시간에 따른 X, Y, Z의 변화를 시각화하여 행동 패턴을 쉽게 이해

```python
# 자이로 X축 데이터의 FFT 적용
gyro_x = gyro_data['X'].values
n = len(gyro_x)
timestep = 0.01  # 가정한 샘플 간격 (단위: 초)

# FFT 적용
fft_gyro_x = fft(gyro_x)
frequencies = np.fft.fftfreq(n, d=timestep)

# 주파수 도메인에서의 그래프 출력
plt.figure(figsize=(10, 5))
plt.plot(frequencies[:n//2], np.abs(fft_gyro_x)[:n//2])  # 절반만 시각화
plt.title('FFT of Gyroscope X')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

### 결과
> 자이로스코프 X축 데이터에 대해 **FFT(푸리에 변환)**를 적용하여 주파수 도메인에서 분석한 결과   
> 시간 도메인에서의 데이터를 주파수 성분으로 변환하여, 주파수별로 신호의 강도를 분석 가능
>   1. 주파수 대역: X축은 `Frequeny (Hz)`를 나타냄, 0에서 50Hz까지의 주파수(각 회전 성분의 진동수), 자이로스코프 데이터는 보통 50Hz의 샘플링 주파수로 수집
>   2. 진폭(Amplitude): Y축은 Amplitude (진폭), 각 주파수 성분의 세기, 진폭이 클수록 해당 주파수에서의 신호가 강하다는 의미
>   3. 주요 주파수 성분: 0Hz에서 매우 큰 진폭이 나타냄, 나머지 주파수에서는 진폭이 급격히 작아지는 경향이 보임, 0Hz에서 `매우 큰 피크: 신호의 DC 성분(즉, 시간에 따라 변하지 않는 상수 성분)이 포함되어 있음을 나타냄`, 자이로스코프의 X축 데이터에 상수 오프셋이 있을 수 있음을 시사. 이 경우, 데이터를 사용하기 전에 DC 오프셋을 제거하는 것이 좋음
>   4. 고주파수 성분의 약함: 0Hz를 제외한 나머지 주파수 성분에서의 진폭이 매우 작음, 고주파 성분이 거의 없다는 것을 의미. 즉, 자이로스코프 X축 데이터는 대부분 저주파 성분으로 구성되어 있으며, 이는 비교적 느린 움직임이나 장기적인 행동 패턴을 나타냄
***
자이로스코프 X축 데이터에서 대부분의 주파수 성분이 0Hz 근처에 집중되어 있고, 고주파 성분은 거의 없는 데이터라는 것을 나타냄, 이는 신호가 주로 저주파수 성분으로 구성되어 있어, 장기적인 움직임이나 느린 행동 변화를 포착하고 있음을 나타냄

### DC 성분 제거
> FFT 분석 결과 0Hz에서 매우 큰 진폭이 나타났으며, DC 오프셋을 제거하는 것이 좋음   
> 이는 데이터를 처리하거나 분석할 때 DC 성분이 신호에 영향을 미치지 않도록 하는 일반적인 방법

```python
# DC 성분 제거
gyro_x_detrended = gyro_x - np.mean(gyro_x)

# 다시 FFT 적용
fft_gyro_x_detrended = fft(gyro_x_detrended)
frequencies = np.fft.fftfreq(n, d=timestep)

# 주파수 도메인에서의 그래프 출력 (DC 성분 제거 후)
plt.figure(figsize=(10, 5))
plt.plot(frequencies[:n//2], np.abs(fft_gyro_x_detrended)[:n//2])
plt.title('FFT of Gyroscope X (DC Offset Removed)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

### 결과
> 1. 0Hz의 DC 성분 제거:
> - 0Hz에서 매우 큰 진폭의 피크를 제거하였고, 저주파수(1~10Hz) 범위에서 신호의 주파수 성분들이 주로 나타남
> 2. 저주파 성분의 존재:
> - 1Hz ~ 10Hz 사이의 저주파 성분이 여전히 높게 나타나고 있음.
> - `이는 자이로스코프 데이터가 주로 느린 움직임이나 천천히 변화하는 행동과 관련된 신호를 포함하고 있음을 시사`
> - `저주파 대역은 보통 사람의 움직임(걷기, 앉기, 서기 등)과 같은 느린 행동과 연관되며, 자이로스코프 신호에서 자주 나타나는 패턴`
> 3. 고주파 성분의 약함
> - 10Hz 이상의 고주파 성분은 거의 존재하지 않으며, 이는 고주파 성분이 미미하거나 신호에 잡음(noise)이 거의 없다는 것을 의미
> - `자이로스코프 데이터를 통해 분석할 때, 대부분의 유의미한 정보는 저주파 영역에서 발생하는 것으로 보임`
***
- 저주파수 주도 신호: DC 성분 제거 후, 여전히 자이로스코프 데이터는 0Hz에 가까운 저주파 성분이 강하게 나타나고 있음, 이 주파수 대역은 주로 사람이 천천히 움직이는 패턴과 연관될 가능성이 큼
- 노이즈 제거: 고주파 성분이 거의 없다는 것은 데이터가 깨긋하고, 주요 신호는 저주파 대역에서 발생한다는 점을 의미
- 추가 분석: 이 주파수 특성을 기반으로 필터링을 적용하거나 특정 주파수 대역을 더욱 강조하여 분석 가능 

### 사람 움직임 패턴에 해당하는 주파수 대역 (1Hz ~ 5Hz) 시각화 코드

```python
# 자이로 X축 데이터의 FFT 적용
gyro_x = gyro_data['X'].values
n = len(gyro_x)
timestep = 0.01  # 샘플 간격 (단위: 초)

# FFT 적용
fft_gyro_x = fft(gyro_x)
frequencies = np.fft.fftfreq(n, d=timestep)

# 양의 주파수 성분만 선택
positive_frequencies = frequencies[:n//2]
positive_amplitude = np.abs(fft_gyro_x)[:n//2]

# 사람이 움직이는 주파수 대역 (1Hz ~ 5Hz)
movement_mask = (positive_frequencies >= 1) & (positive_frequencies <= 5)

# 사람이 움직이는 주파수 대역만 시각화
plt.figure(figsize=(10, 5))
plt.plot(positive_frequencies[movement_mask], positive_amplitude[movement_mask])
plt.title('Movement-related Frequency Range (1Hz - 5Hz) of Gyroscope X')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

> 1. 주요 진폭 성분:
>  - 주파수 1Hz에서 3.5Hz 구간에서 강한 진폭 성분이 나타남, 이는 주로 사람의 동작(걷기, 서기, 앉기 등)이 이 주파수 대역에서 발생한다는 것을 의미
>  - 이 범위는 걷기와 같은 주기적인 동작이 발생하는 주파수 대역이며, 진폭이 높은 부분은 이러한 행동이 빈번히 발생했음을 나타냄
> 2. 주파수 대역:
>  - 1Hz 이하의 주파수에서는 진폭이 상대적으로 낮으며, 이는 주로 저속 움직임(매우 느리거나 거의 움직이지 않는 상태)을 나타냄
>  - 1Hz ~ 5Hz 범위에서 자이로스코프 신호가 상당히 큰 진폭을 보이므로, 사람이 움직이는 동안 발생하는 주기적인 회전이나 동작을 반영하고 있음을 확인
> 3. 진폭의 감소:
>  - 주파수가 4Hz를 넘어서면서 진폭이 서서히 감소하는 경향을 보임, 이는 이 구간에서 더 빠른 움직임이 있거나, 해당 대역에서의 움직임이 상대적을 적다는 것을 의미
***
- 1Hz ~ 3.5Hz 구간에서 가장 강한 진폭을 보이며, 이는 사람이 걷기, 서기, 앉기 등의 동작과 연관된 주파수 대역
- 3.5Hz 이후에는 진폭이 점차 감소하므로, 이 구간에서는 더 빠르거나 미약한 움직임이 있을 수 있음
- 분석 가능성: 이 주파수 대역을 기준으로 움직임이 빈번하게 발생하는 패턴을 더 자세히 분석 가능, 특정 행동과 연관된 주파수를 확인 가능

### 전체 데이터셋에서 X, Y, Z 축 모두 푸리에 변환
> X, Y, Z 축에 대해 모두 푸리에 변환을 적용하여 1Hz ~ 5Hz 범위에서의 주파수 성분을 분석

```python
# 전체 데이터셋에서 X, Y, Z 축 데이터 가져오기
gyro_x = gyro_data['X'].values
gyro_y = gyro_data['Y'].values
gyro_z = gyro_data['Z'].values
n = len(gyro_x)
timestep = 0.01  # 샘플 간격 (단위: 초)

# X, Y, Z축에 대해 FFT 적용
fft_gyro_x = fft(gyro_x)
fft_gyro_y = fft(gyro_y)
fft_gyro_z = fft(gyro_z)

# 양의 주파수 성분만 선택
frequencies = np.fft.fftfreq(n, d=timestep)
positive_frequencies = frequencies[:n//2]
amplitude_x = np.abs(fft_gyro_x)[:n//2]
amplitude_y = np.abs(fft_gyro_y)[:n//2]
amplitude_z = np.abs(fft_gyro_z)[:n//2]

# 사람이 움직이는 주파수 대역 (1Hz ~ 5Hz)
movement_mask = (positive_frequencies >= 1) & (positive_frequencies <= 5)

# X, Y, Z 축의 주파수 성분 시각화
plt.figure(figsize=(15, 5))

plt.subplot(3, 1, 1)
plt.plot(positive_frequencies[movement_mask], amplitude_x[movement_mask], label='X')
plt.title('Movement-related Frequency (1Hz - 5Hz) for X-axis (Entire Dataset)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(positive_frequencies[movement_mask], amplitude_y[movement_mask], label='Y', color='g')
plt.title('Movement-related Frequency (1Hz - 5Hz) for Y-axis (Entire Dataset)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(positive_frequencies[movement_mask], amplitude_z[movement_mask], label='Z', color='b')
plt.title('Movement-related Frequency (1Hz - 5Hz) for Z-axis (Entire Dataset)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

### 결과
> 1. X축 (청록색 그래프)
>   - 패턴: X축에서는 1Hz ~ 2Hz 대역에서 진폭이 상대적으로 높고, 그 이후로는 진폭이 점차 감소하는 패턴을 보임, 이는 좌우 방향의 움직임에서 주로 1Hz ~ 2Hz 대역에서 활동이 나타나고 있음을 시사
>   - 의미: 이 축에서의 주요 주파수 성분은 `걷기` 또는 `좌우 움직임`과 관련이 있을 가능성이 큼, 사람의 걷기 동작은 주로 이 주파수 대역에서 발생하므로, X축에서의 움직임이 이 대역에서 뚜렷하게 나타나는 것임
> 2. Y축 (초록색 그래프)
>   - 패턴: Y축에서는 1Hz ~ 2.5Hz 범위에서 진폭이 높고, 그 이후로는 서서히 감소하는 패턴을 보여줌, 이 대역은 X축보다 약간 더 넓은 주파수 대역에서 높은 진폭을 나타냄
>   - 의미: Y축은 주로 위아래 방향의 움직임을 측정하므로, 이 주파수 대역은 `걷기`, `뛰기`와 같은 `상하` 움직임에서 발생, 이 패턴은 `사람이 보통 걷거나 뛰는 동안 위아래로 발생하는 회전 운동`을 나타냄
> 3. Z축 (파란색 그래프)
>   - 패턴: Z축에서는 1Hz ~ 2Hz 대역에서 진폭이 높으며, 이후 3Hz까지 비교적 일정한 진폭을 유지하다가 서서히 감소하는 경향이 보임, 다른 축과 유사하게 저주파 대역에서 강한 성분이 나타남
>   - 의미: Z축은 `앞뒤 방향`의 움직임을 반영하므로, 이 패턴은 주로 `앞으로 걷거나 움직이는 동안 발생`하는 주파수 성분
***
1. 주요 주파수 대역:
  - X, Y, Z 축 모두에서 1Hz ~ 2Hz 범위에서 가장 강한 진폭이 나타남
  - 이 주파수 대역은 걷기, 서기, 앉기 등의 반복적인 사람의 움직임과 관련이 있음
  - 보통 사람의 걷는 주기나 움직임의 반복 패턴이 나타나는 주파수 범위
2. 진폭의 감소:
  - 각 축 모두 2Hz 이후부터 진폭이 점차 감소하는 경향을 보임
  - 이는 2Hz 이상의 고주파 대역에서는 사람이 움직이는 행동이 덜 발생하거나, 미세한 움직임만 반영될 수 있음을 나타냄
3. 축 간 비교:
  - Y축에서 가장 높은 진폭이 나타나며, 이는 사람이 걷거나 뛸 때 위아래로 발생하는 움직임이 상대적으로 크기 때문일 수 있음
  - X축과 Z축에서는 상대적으로 Y축보다 낮은 진폭을 보이지만, 여전히 일정한 패턴을 나타내며, 각각 `좌우` 또는 `앞뒤` 움직임을 반영하고 있을 가능성이 큼
