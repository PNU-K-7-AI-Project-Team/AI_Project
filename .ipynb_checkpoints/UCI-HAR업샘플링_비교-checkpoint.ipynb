1. 데이터 로드: UCI-HAR 데이터셋을 로드합니다.
2. 업샘플링: UCI-HAR 데이터의 주파수를 자이로 데이터의 주파수와 맞추기 위해 업샘플링합니다.
3. 시각화: 업샘플링된 UCI-HAR 데이터와 자이로 데이터를 시각적으로 비교합니다.
1. 데이터 로드
먼저 UCI-HAR 데이터셋과 자이로 데이터를 불러옵니다.
2. 업샘플링
UCI-HAR 데이터셋이 50Hz로 수집되었기 때문에, 자이로 데이터가 100Hz라면 UCI-HAR 데이터를 두 배로 업샘플링해야 합니다.
선형 보간법을 사용하여 두 데이터셋의 샘플링 주기를 일치시킵니다.
3. 시각화
두 데이터셋을 같은 주파수로 맞춘 후, 각 축(X, Y, Z)에 대해 시계열 데이터를 시각화하여 비교합니다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# UCI-HAR 데이터 로드 (X, Y, Z 축 데이터)
uci_har_path = './원본 데이터/UCI HAR Dataset/'  # 실제 경로로 수정
gyro_x_train = pd.read_csv(uci_har_path + 'train/Inertial Signals/body_gyro_x_train.txt', sep='\s+', header=None).values
gyro_y_train = pd.read_csv(uci_har_path + 'train/Inertial Signals/body_gyro_y_train.txt', sep='\s+', header=None).values
gyro_z_train = pd.read_csv(uci_har_path + 'train/Inertial Signals/body_gyro_z_train.txt', sep='\s+', header=None).values

# UCI-HAR 데이터는 50Hz, 자이로 데이터는 100Hz라고 가정
current_freq = 50  # UCI-HAR 주파수
desired_freq = 100  # 자이로 데이터 주파수

# 시간 축 생성 (50Hz UCI-HAR 데이터에 맞춤)
t_current = np.linspace(0, len(gyro_x_train)/current_freq, num=len(gyro_x_train))

# 새로운 시간 축 생성 (100Hz 자이로 데이터에 맞춰 업샘플링)
t_new = np.linspace(0, len(gyro_x_train)/current_freq, num=len(gyro_x_train)*2)

# X축 선형 보간을 통한 업샘플링
interp_x = interp1d(t_current, gyro_x_train.mean(axis=1), kind='linear')
uci_x_upsampled = interp_x(t_new)

# Y축 선형 보간을 통한 업샘플링
interp_y = interp1d(t_current, gyro_y_train.mean(axis=1), kind='linear')
uci_y_upsampled = interp_y(t_new)

# Z축 선형 보간을 통한 업샘플링
interp_z = interp1d(t_current, gyro_z_train.mean(axis=1), kind='linear')
uci_z_upsampled = interp_z(t_new)

# 시각화
plt.figure(figsize=(14, 8))

# X축 시각화
plt.subplot(3, 1, 1)
plt.plot(t_new, uci_x_upsampled, label="UCI-HAR X (Upsampled)", color='b')
plt.title('X-axis: Upsampled UCI-HAR Data (Linear Interpolation)')
plt.legend()

# Y축 시각화
plt.subplot(3, 1, 2)
plt.plot(t_new, uci_y_upsampled, label="UCI-HAR Y (Upsampled)", color='b')
plt.title('Y-axis: Upsampled UCI-HAR Data (Linear Interpolation)')
plt.legend()

# Z축 시각화
plt.subplot(3, 1, 3)
plt.plot(t_new, uci_z_upsampled, label="UCI-HAR Z (Upsampled)", color='b')
plt.title('Z-axis: Upsampled UCI-HAR Data (Linear Interpolation)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from scipy.interpolate import interp1d

# 자이로 데이터 로드 (RegisterDate를 datetime으로 변환)
gyro_data = pd.read_csv('./원본 데이터/자이로 데이터.csv')  # 실제 경로로 수정
gyro_data['RegisterDate'] = pd.to_datetime(gyro_data['RegisterDate'])

# UCI-HAR 데이터 로드
uci_har_path = './원본 데이터/UCI HAR Dataset/'  # 실제 경로로 수정
gyro_x_train = pd.read_csv(uci_har_path + 'train/Inertial Signals/body_gyro_x_train.txt', sep='\s+', header=None).values
gyro_y_train = pd.read_csv(uci_har_path + 'train/Inertial Signals/body_gyro_y_train.txt', sep='\s+', header=None).values
gyro_z_train = pd.read_csv(uci_har_path + 'train/Inertial Signals/body_gyro_z_train.txt', sep='\s+', header=None).values

# X, Y, Z 축 데이터를 하나의 DataFrame으로 병합
uci_har_gyro_df = pd.DataFrame({
    'X': gyro_x_train.mean(axis=1),
    'Y': gyro_y_train.mean(axis=1),
    'Z': gyro_z_train.mean(axis=1)
})

# 자이로 데이터를 UCI-HAR 데이터의 길이로 슬라이싱
n_samples_uci = len(uci_har_gyro_df)
gyro_sliced = gyro_data.iloc[:n_samples_uci]  # UCI-HAR 데이터 길이만큼 자이로 데이터를 슬라이싱

# 데이터 정규화 (0~1 사이로)
scaler = MinMaxScaler()

gyro_sliced[['X', 'Y', 'Z']] = scaler.fit_transform(gyro_sliced[['X', 'Y', 'Z']])
uci_har_gyro_df[['X', 'Y', 'Z']] = scaler.fit_transform(uci_har_gyro_df[['X', 'Y', 'Z']])

# 보간 방식 리스트
methods = ['linear', 'nearest', 'cubic']

plt.figure(figsize=(14, 20))

# X, Y, Z축에 대해 보간 및 시각화
for i, method in enumerate(methods):
    # X축 보간
    interp_x = interp1d(np.arange(len(uci_har_gyro_df['X'])), uci_har_gyro_df['X'], kind=method)
    uci_x_upsampled = interp_x(np.linspace(0, len(uci_har_gyro_df['X'])-1, num=len(gyro_sliced)))

    # Y축 보간
    interp_y = interp1d(np.arange(len(uci_har_gyro_df['Y'])), uci_har_gyro_df['Y'], kind=method)
    uci_y_upsampled = interp_y(np.linspace(0, len(uci_har_gyro_df['Y'])-1, num=len(gyro_sliced)))

    # Z축 보간
    interp_z = interp1d(np.arange(len(uci_har_gyro_df['Z'])), uci_har_gyro_df['Z'], kind=method)
    uci_z_upsampled = interp_z(np.linspace(0, len(uci_har_gyro_df['Z'])-1, num=len(gyro_sliced)))

    # X축 시각화
    plt.subplot(len(methods), 3, i * 3 + 1)
    plt.plot(uci_x_upsampled, label=f"UCI-HAR X (Upsampled - {method})", color='b')
    plt.plot(gyro_sliced['X'], label="Gyro X (Sliced)", color='r', alpha=0.6)
    plt.title(f'X-axis: {method} interpolation')
    plt.legend()

    # Y축 시각화
    plt.subplot(len(methods), 3, i * 3 + 2)
    plt.plot(uci_y_upsampled, label=f"UCI-HAR Y (Upsampled - {method})", color='b')
    plt.plot(gyro_sliced['Y'], label="Gyro Y (Sliced)", color='r', alpha=0.6)
    plt.title(f'Y-axis: {method} interpolation')
    plt.legend()

    # Z축 시각화
    plt.subplot(len(methods), 3, i * 3 + 3)
    plt.plot(uci_z_upsampled, label=f"UCI-HAR Z (Upsampled - {method})", color='b')
    plt.plot(gyro_sliced['Z'], label="Gyro Z (Sliced)", color='r', alpha=0.6)
    plt.title(f'Z-axis: {method} interpolation')
    plt.legend()

plt.tight_layout()
plt.show()
```
