```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 로드
file_path = './DATA/gyro_data.csv'
data = pd.read_csv(file_path)

# 라벨을 숫자로 변환
labels = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}

# 데이터 전처리
data['classe'] = data['classe'].map(labels)
data = data.dropna()

# x, y, z 데이터 정규화
mean = np.mean(data[['gyros_forearm_x', 'gyros_forearm_y', 'gyros_forearm_z']], axis=0)
std = np.std(data[['gyros_forearm_x', 'gyros_forearm_y', 'gyros_forearm_z']], axis=0)
data[['gyros_forearm_x', 'gyros_forearm_y', 'gyros_forearm_z']] = (data[['gyros_forearm_x', 'gyros_forearm_y', 'gyros_forearm_z']] - mean) / std

# 데이터 분할 (훈련 데이터와 테스트 데이터)
train_size = int(0.7 * len(data))
train_data = data[:train_size]
test_data = data[train_size:]

x_train = train_data[['gyros_forearm_x', 'gyros_forearm_y', 'gyros_forearm_z']].values
y_train = train_data['classe'].values

x_test = test_data[['gyros_forearm_x', 'gyros_forearm_y', 'gyros_forearm_z']].values
y_test = test_data['classe'].values

# 데이터 세그먼트화 및 라벨링
TIME_PERIODS = 80  # 세그먼트 길이
STEP_DISTANCE = 40 # if STEP_DISTANCE = TIME_PERIODS, then no overlap

def data_segments(x_data, y_data):
    segments = []
    labels = []
    for i in range(0, len(x_data) - TIME_PERIODS, STEP_DISTANCE):
        X = x_data[i:i + TIME_PERIODS]
        label = y_data[i + TIME_PERIODS - 1]  # 세그먼트 마지막의 라벨 선택
        segments.append(X)
        labels.append(label)
    
    # reshape (-1, TIME_PERIODS, 3)
    segments = np.array(segments, dtype=np.float32).reshape(-1, TIME_PERIODS, 3)
    labels = np.asarray(labels)
    return segments, labels

# 훈련 및 테스트 데이터 세그먼트화
x_train, y_train = data_segments(x_train, y_train) 	
x_test, y_test = data_segments(x_test, y_test)	
print("x_train.shape=", x_train.shape)
print("x_test.shape=",  x_test.shape)

# 라벨 원-핫 인코딩 
y_train = tf.keras.utils.to_categorical(y_train, num_classes=5)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=5)

print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

# 1D CNN 모델 생성
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=(TIME_PERIODS, 3)))  # shape=(80,3)
model.add(tf.keras.layers.Conv1D(filters=100, kernel_size=11, activation='relu'))
model.add(tf.keras.layers.MaxPool1D())
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Conv1D(filters=10, kernel_size=5, activation='relu'))
model.add(tf.keras.layers.MaxPool1D())
model.add(tf.keras.layers.Dropout(rate=0.5))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(units=5, activation='softmax'))  # 5개의 클래스 (A-E)
model.summary()

# 모델 컴파일 및 학습
opt = tf.keras.optimizers.RMSprop(learning_rate=0.001)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
ret = model.fit(x_train, y_train, epochs=100, batch_size=400, validation_data=(x_test, y_test), verbose=2)

# 모델 평가
train_loss, train_acc = model.evaluate(x_train, y_train, verbose=2)
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)

# 정확도 및 손실 시각화
plt.title("Accuracy")
plt.plot(ret.history['accuracy'], "b-", label="train accuracy")
plt.plot(ret.history['val_accuracy'], "r-", label="val accuracy")
plt.plot(ret.history['loss'], "g-", label="train loss")
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend(loc="best")
plt.show()

train_acc, test_acc
```
