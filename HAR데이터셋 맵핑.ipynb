# **1. 데이터 로드 및 전처리**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 데이터셋 로드
transformed_data = pd.read_csv('./원본 데이터/transformed_train.csv')

# 파일을 한 줄씩 읽어서 수동으로 처리
rows = []
with open('./휴대폰 GPS, 자이로센서 분석 인간 행동 인식 식별/DATA/WISDM_ar_v1.1/WISDM_ar_v1.1_raw.txt') as file:
    for line in file:
        parts = line.strip().split(',')
        if len(parts) == 6:
            rows.append(parts)

# DataFrame으로 변환
wisdm_data = pd.DataFrame(rows, columns=['user', 'activity', 'timestamp', 'x-accl', 'y-accl', 'z-accl'])

# 빈 문자열 또는 비정상적인 데이터를 NaN으로 대체
wisdm_data.replace('', np.nan, inplace=True)

# NaN이 포함된 행 제거
wisdm_data.dropna(inplace=True)

# z-accl 열을 float으로 변환
wisdm_data['z-accl'] = wisdm_data['z-accl'].str.replace(';', '').astype(float)
wisdm_data['x-accl'] = wisdm_data['x-accl'].astype(float)
wisdm_data['y-accl'] = wisdm_data['y-accl'].astype(float)

# 데이터가 올바르게 처리되었는지 확인
print(wisdm_data.head())

# 각 데이터셋의 클래스별 통계적 특징 계산
def calculate_statistics(df):
    stats = df.groupby('activity').agg({
        'x-accl': ['mean', 'std', 'min', 'max'],
        'y-accl': ['mean', 'std', 'min', 'max'],
        'z-accl': ['mean', 'std', 'min', 'max']
    })
    return stats

# 통계적 특징 계산
transformed_stats = calculate_statistics(transformed_data)
wisdm_stats = calculate_statistics(wisdm_data)

# 통계적 특징 출력
print("Transformed Dataset Statistics:")
print(transformed_stats)
print("\nWISDM Dataset Statistics:")
print(wisdm_stats)


# 시각화를 통해 각 클래스의 가속도 분포를 비교
def plot_acceleration_distribution(data, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 3, 1)
    plt.hist(data['x-accl'], bins=50, alpha=0.7, label='x-accl')
    plt.title(f'{title} - x-accl')
    plt.subplot(1, 3, 2)
    plt.hist(data['y-accl'], bins=50, alpha=0.7, label='y-accl')
    plt.title(f'{title} - y-accl')
    plt.subplot(1, 3, 3)
    plt.hist(data['z-accl'], bins=50, alpha=0.7, label='z-accl')
    plt.title(f'{title} - z-accl')
    plt.show()

# transformed_train.csv 각 클래스의 가속도 분포
for activity in transformed_data['activity'].unique():
    plot_acceleration_distribution(transformed_data[transformed_data['activity'] == activity], f'transformed_train.csv - {activity}')

# WISDM 각 클래스의 가속도 분포
for activity in wisdm_data['activity'].unique():
    plot_acceleration_distribution(wisdm_data[wisdm_data['activity'] == activity], f'WISDM - {activity}')


# 클래스 매핑 (통계적 특징을 바탕으로)
class_mapping = {
    'A': 'Walking',
    'B': 'Jogging',
    'C': 'Upstairs',
    'D': 'Downstairs',
    'E': 'Sitting'
}

# transformed_train.csv의 클래스(A, B, C, D, E)를 매핑된 클래스로 변경
transformed_data['activity'] = transformed_data['activity'].map(class_mapping)

# 결과 확인
print("Mapped Transformed Dataset Class Distribution:")
print(transformed_data['activity'].value_counts())

# 변경된 데이터를 새로운 CSV 파일로 저장
transformed_data.to_csv('./원본 데이터/transformed_train_mapped.csv', index=False)
```
